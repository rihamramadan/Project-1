```python
%matplotlib 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
client_report = "Resources/ClientReport.csv"
ambulatory_service = "Resources/ClientReportAmbulatoryService.csv"
cd4_test = "Resources/ClientReportCd4Test.csv"
hispanic_subgroup = "Resources/ClientReportHispanicSubgroup.csv"
risk_factor = "Resources/ClientReportHivRiskFactor.csv"
medical_insurance = "Resources/ClientReportMedicalInsurance.csv"
nhpi_subgroup = "Resources/ClientReportNhpiSubgroup.csv"
asian_subgroup = "Resources/ClientReportAsianSubgroup.csv"
race = "Resources/ClientReportRace.csv"
support_services = "Resources/ClientReportServiceDelivered.csv"
service_visit = "Resources/ClientReportServiceVisits.csv"
viral_load = "Resources/ClientReportViralLoadTest.csv"


client_report_df = pd.read_csv(client_report)
ambulatory_service_df = pd.read_csv(ambulatory_service)
cd4_test_df = pd.read_csv(cd4_test)
hispanic_subgroup_df = pd.read_csv(hispanic_subgroup)
risk_factor_df = pd.read_csv(risk_factor)
medical_insurance_df = pd.read_csv(medical_insurance)
nhpi_subgroup_df = pd.read_csv(nhpi_subgroup)
asian_subgroup_df = pd.read_csv(asian_subgroup)
race_df = pd.read_csv(race)
support_services_df = pd.read_csv(support_services)
service_visit_df = pd.read_csv(service_visit)
viral_load_df = pd.read_csv(viral_load)
```

```python
hispanic_subgroup_df = hispanic_subgroup_df.replace({'HispanicSubgroupId': 1}, 'Mexican')    
hispanic_subgroup_df = hispanic_subgroup_df.replace({'HispanicSubgroupId': 2}, 'Puerto Rican')    
hispanic_subgroup_df = hispanic_subgroup_df.replace({'HispanicSubgroupId': 3}, 'Cuban')    
hispanic_subgroup_df = hispanic_subgroup_df.replace({'HispanicSubgroupId': 4}, 'Other')    
   
hispanic_subgroup_df.head()
```

```python
risk_factor_df = risk_factor_df.replace({'HivRiskFactorId': 1}, 'MSM')    
risk_factor_df = risk_factor_df.replace({'HivRiskFactorId': 2}, 'IDU')    
risk_factor_df = risk_factor_df.replace({'HivRiskFactorId': 3}, 'Hemophilia/coagulation disorder')    
risk_factor_df = risk_factor_df.replace({'HivRiskFactorId': 4}, 'Heterosexual contact')    
risk_factor_df = risk_factor_df.replace({'HivRiskFactorId': 5}, 'Blood Transfusion')    
risk_factor_df = risk_factor_df.replace({'HivRiskFactorId': 6}, 'Perinatal Transmission')    
risk_factor_df = risk_factor_df.replace({'HivRiskFactorId': 9}, 'Unknown')    


risk_factor_df.head()
```

```python
medical_insurance_df = medical_insurance_df.replace({'MedicalInsuranceId': 10}, 'Employer')    
medical_insurance_df = medical_insurance_df.replace({'MedicalInsuranceId': 11}, 'Individual')    
medical_insurance_df = medical_insurance_df.replace({'MedicalInsuranceId': 2}, 'Medicare')    
medical_insurance_df = medical_insurance_df.replace({'MedicalInsuranceId': 12}, 'Medicaid')    
medical_insurance_df = medical_insurance_df.replace({'MedicalInsuranceId': 13}, 'VA')    
medical_insurance_df = medical_insurance_df.replace({'MedicalInsuranceId': 14}, 'IHS')    
medical_insurance_df = medical_insurance_df.replace({'MedicalInsuranceId': 15}, 'Other')    
medical_insurance_df = medical_insurance_df.replace({'MedicalInsuranceId': 16}, 'No insurance/uninsured')    

medical_insurance_df.head()
```

```python
nhpi_subgroup_df= nhpi_subgroup_df.replace({'NhpiSubgroupId': 1}, 'Native Hawaiian')    
nhpi_subgroup_df= nhpi_subgroup_df.replace({'NhpiSubgroupId': 2}, 'Guamanian or Chamorro')    
nhpi_subgroup_df= nhpi_subgroup_df.replace({'NhpiSubgroupId': 3}, 'Samoan')    
nhpi_subgroup_df= nhpi_subgroup_df.replace({'NhpiSubgroupId': 4}, 'Other Pacific Islander')    

nhpi_subgroup_df.head()
```

```python
asian_subgroup_df = asian_subgroup_df.replace({'AsianSubgroupId': 1}, 'Asian Indian')
asian_subgroup_df = asian_subgroup_df.replace({'AsianSubgroupId': 2}, 'Chinese')
asian_subgroup_df = asian_subgroup_df.replace({'AsianSubgroupId': 3}, 'Filipino')
asian_subgroup_df = asian_subgroup_df.replace({'AsianSubgroupId': 4}, 'Japanese')
asian_subgroup_df = asian_subgroup_df.replace({'AsianSubgroupId': 5}, 'Korean')
asian_subgroup_df = asian_subgroup_df.replace({'AsianSubgroupId': 6}, 'Vietnamese')
asian_subgroup_df = asian_subgroup_df.replace({'AsianSubgroupId': 7}, 'Other')

asian_subgroup_df.head()
```

```python
race_df = race_df.replace({'RaceId': 1}, 'White')
race_df = race_df.replace({'RaceId': 2}, 'African American')
race_df = race_df.replace({'RaceId': 3}, 'Asian')
race_df = race_df.replace({'RaceId': 4}, ' Native Hawaiian/Pacific Islander')
race_df = race_df.replace({'RaceId': 5}, 'American Indian or Alaska Native')

race_df.head()
```

```python
support_services_df = support_services_df.replace({'ServiceID': 32}, 'Psychosocial Support')

support_services_df.head()
```

```python
service_visit_df = service_visit_df.replace({'ServiceID': 8}, 'Medical ')
service_visit_df = service_visit_df.replace({'ServiceID': 10}, 'Dental')
service_visit_df = service_visit_df.replace({'ServiceID': 11}, 'EIS')
service_visit_df = service_visit_df.replace({'ServiceID': 16}, 'Mental Health')
service_visit_df = service_visit_df.replace({'ServiceID': 18}, 'Medical Case Management')
service_visit_df = service_visit_df.replace({'ServiceID': 19}, 'Substance Use')

service_visit_df.head()
```

```python
merged_df = pd.merge(client_report_df, hispanic_subgroup_df, how ='outer', on='ClientId')
merged_df = pd.merge(merged_df, ambulatory_service_df, how ='outer', on='ClientId')
merged_df = pd.merge(merged_df, cd4_test_df, how ='outer', on='ClientId')
merged_df = pd.merge(merged_df, risk_factor_df, how ='outer', on='ClientId')
merged_df = pd.merge(merged_df, medical_insurance_df, how ='outer', on='ClientId')
merged_df = pd.merge(merged_df, race_df, how ='outer', on='ClientId')
merged_df = pd.merge(merged_df, support_services_df, how ='outer', on='ClientId')
merged_df = pd.merge(merged_df, service_visit_df, how ='outer', on='ClientId')
merged_df = pd.merge(merged_df, viral_load_df, how ='outer', on='ClientId')


merged_df.head()
```

```python
master_df = master_df.rename(columns={"ClientDateofBirth":"Birthdate",
                                      "EthnicityID": "Ethnicity",
                                      "GenderID": "Gender",
                                      "PovertyLevelID": "Poverty Level",
                                      "HousingStatusID": "Housing Status",
                                      "HivAidsStatusID": "HIV/AIDS Status",
                                      "PrescribedArtID": "Prescribed ARTs",
                                      "ScreenedTBSinceHivDiagnosisID": "Screened TB", 
                                      "ScreenedSyphilisID": "Screened Syphilis",
                                      "ScreenedHepatitisBSinceHivDiagnosisID":"Screened Hepatitis B",
                                      "VaccinatedHepatitisBID":"Vaccinated Hepatitis B",
                                      "ScreenedHepatitisCSinceHivDiagnosisID":"Screened Hepatitis C",
                                      "SexAtBirthID":"Sex At Birth",
                                      "HivDiagnosisYear": "HIV Diagnosis Year",
                                      "HivPosTestDate": "Positive Test Date",
                                      "OamcLinkDate": "Linked to Care Date",
                                      "HispanicSubgroupId":"Hispanic Subgroup",
                                      "ServiceDate_x":"Cd4 Count Date",
                                      "ServiceDate_y":"Viral Load Test Date",
                                      "Count_x":"Cd4 Test Value",
                                      "Count_y":"Viral Load Value",
                                      "HivRiskFactorId": "HIV Risk Factor",
                                      "MedicalInsuranceId": "Medical Insurance",
                                      "RaceId": "Race",
                                      "ServiceID_x":"Support Services",
                                      "ServiceID_y":"Core Services"
                                     
                                     })

master_df.columns
```

```python
master_df = master_df.replace({'Ethnicity': 1}, 'Hispanic')
master_df = master_df.replace({'Ethnicity': 2}, 'Non-Hispanic')

master_df = master_df.replace({'Gender': 1}, 'Male')
master_df = master_df.replace({'Gender': 2}, 'Female')
master_df = master_df.replace({'Gender': 4}, 'Unknown')
master_df = master_df.replace({'Gender': 6}, 'Transfemale')
master_df = master_df.replace({'Gender': 7}, 'Transmale')
master_df = master_df.replace({'Gender': 8}, 'Trangender Other')

master_df = master_df.replace({'Poverty Level': 13}, 'Below 100%')
master_df = master_df.replace({'Poverty Level': 9}, '100-138%')
master_df = master_df.replace({'Poverty Level': 10}, '139%-200%')
master_df = master_df.replace({'Poverty Level': 11}, '201-250%')
master_df = master_df.replace({'Poverty Level': 12}, '251-400%')
master_df = master_df.replace({'Poverty Level': 7}, '401-500%')
master_df = master_df.replace({'Poverty Level': 8}, '>500%')

master_df = master_df.replace({'Housing Status': 1}, 'Stable/permanent')
master_df = master_df.replace({'Housing Status': 2}, 'Temporary')
master_df = master_df.replace({'Housing Status': 3}, 'Unstable')

master_df = master_df.replace({'HIV/AIDS Status': 1}, 'HIV negative')
master_df = master_df.replace({'HIV/AIDS Status': 2}, 'HIV-positive, not AIDS')
master_df = master_df.replace({'HIV/AIDS Status': 3}, 'HIV-positive, AIDS status unknown')
master_df = master_df.replace({'HIV/AIDS Status': 4}, 'CDC-defined AIDS')
master_df = master_df.replace({'HIV/AIDS Status': 27}, 'HIV indeterminate')


master_df = master_df.replace({'Prescribed ARTs': 1}, 'Yes')
master_df = master_df.replace({'Prescribed ARTs': 7}, 'No')

master_df = master_df.replace({'Screened TB': 1}, 'No')
master_df = master_df.replace({'Screened TB': 2}, 'Yes')
master_df = master_df.replace({'Screened TB': 3}, 'Not medically indicated')
master_df = master_df.replace({'Screened TB': 4}, 'Unknown')

master_df = master_df.replace({'Screened Syphilis': 1}, 'No')
master_df = master_df.replace({'Screened Syphilis': 2}, 'Yes')
master_df = master_df.replace({'Screened Syphilis': 3}, 'Not medically indicated')

master_df = master_df.replace({'Screened Hepatitis B': 1}, 'No')
master_df = master_df.replace({'Screened Hepatitis B': 2}, 'Yes')
master_df = master_df.replace({'Screened Hepatitis B': 3}, 'Not medically indicated')
master_df = master_df.replace({'Screened Hepatitis B': 4}, 'Unknown')

master_df = master_df.replace({'Vaccinated Hepatitis B': 1}, 'No')
master_df = master_df.replace({'Vaccinated Hepatitis B': 2}, 'Yes')
master_df = master_df.replace({'Vaccinated Hepatitis B': 3}, 'Not medically indicated')


master_df = master_df.replace({'Screened Hepatitis C': 1}, 'No')
master_df = master_df.replace({'Screened Hepatitis C': 2}, 'Yes')
master_df = master_df.replace({'Screened Hepatitis C': 3}, 'Not medically indicated')
master_df = master_df.replace({'Screened Hepatitis C': 4}, 'Unknown')

master_df = master_df.replace({'Sex At Birth': 1}, 'Male')
master_df = master_df.replace({'Sex At Birth': 2}, 'Female')

```

```python
visits_clean_df = master_df.loc[:, ["ClientId", "Gender","Core Services", "Visits"]]

visits_clean_df.drop_duplicates(inplace=True)
```

```python
gender_visits = visits_clean_df.groupby("Gender")["Visits"].sum()
total_visits = sum(gender_visits)
gender_percentage = 100 * gender_visits/ total_visits 
plt.pie(gender_percentage, 
        labels=["Female", "Male", "Transfemale", "Transmale", "Unknown"], 
        colors=["lightskyblue","lightcoral","lightblue", "purple","orange"], 
        explode=[0, 0.2,0,0,0], 
        autopct='%1.1f%%', 
        shadow=False, startangle=200)


plt.legend(loc="best")

plt.title("Percent of Total Visits by Gender Identity")

fig = plt.gcf()
fig.set_size_inches(9,9)

plt.show()
```

```python
gender_count = master_df.groupby("Gender")["ClientId"].nunique()

x_axis = "Female", "Male", "Transfemale", "Transmale", "Unknown"

plt.bar(x_axis, gender_visits, color="blue", 
        alpha=0.5, align="center")

plt.title("Count of Patients by Gender Identity")
plt.xlabel("Gender Identity")
plt.ylabel("Total Number of Visits")


plt.show()
```

```python
viral_load_df = master_df.loc[:, ["ClientId", "Viral Load Test Date","Viral Load Value","Core Services","Visits"]]
viral_load_dedupe_df = viral_load_df.sort_values(by=['Viral Load Test Date']).drop_duplicates(subset='ClientId', keep='last')
viral_cleaned_df = viral_load_dedupe_df[~viral_load_dedupe_df['Viral Load Value'].isin(['NaN'])]

viral_cleaned_df.head()
```

```python
viral_load = viral_cleaned_df.groupby('ClientId').mean()['Viral Load Value']

visits = viral_cleaned_df.groupby('ClientId').sum()['Visits']

plt.scatter(viral_load, visits,
            s=300, c="coral", 
            edgecolor="black", linewidths=.5, marker="o", 
            alpha=1.0, label="Viral Load Suppression")

plt.title("Viral Load Test Value vs. Number of Visits")
plt.ylabel("Visit Sum by Client")
plt.xlabel("Test Value")
plt.xlim((15,1000000))
plt.grid(True)

plt.gcf().set_size_inches((10, 10))    

plt.show()
```

```python
cd4_load_df = master_df.loc[:, ["ClientId", "Cd4 Count Date","Cd4 Test Value","Core Services","Visits"]]
cd4_dedupe_df = cd4_load_df.sort_values(by=['Cd4 Count Date']).drop_duplicates(subset='ClientId', keep='last')
cd4_cleaned_df = cd4_dedupe_df[~cd4_dedupe_df['Cd4 Test Value'].isin(['NaN'])]

cd4_cleaned_df.head()
```

```python
Cd4 = cd4_cleaned_df.groupby('ClientId')['Cd4 Test Value'].mean()
visits = cd4_cleaned_df.groupby('ClientId').mean()['Visits']

plt.scatter(Cd4,visits, 
            s=200, c="skyblue", 
            edgecolor="black", linewidths=.5, marker="o", 
            alpha=1.0, label="Cd4 Value")

plt.title("Cd4 Value vs. Average Visit")
plt.ylabel("Visit Average by Client")
plt.xlabel("Test Value")
plt.xlim((15,1500))
plt.grid(True)

plt.gcf().set_size_inches((10, 10))    

plt.show()
```
